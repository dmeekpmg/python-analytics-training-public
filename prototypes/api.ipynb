```python
import os
from dotenv import load_dotenv
import requests
```

```python
BASE_URL = "https://api.transport.nsw.gov.au"
```

```python
app_name = os.getenv("APP_NAME")
api_key = os.getenv("API_KEY")
```

```python
API_URI = f"{BASE_URL}/v1/gtfs/vehiclepos/buses"
```

```python
API_URI
```

```python
headers = {
    "Authorization": f"apikey {api_key}"
}
response = requests.get(
    API_URI,
    headers=headers,
    verify='..\\caadmin.netskope.com.pem', 
    stream=True
)
```

```python
response.content.decode('unicode_escape')
```

```python
from google.transit import gtfs_realtime_pb2
import requests
from google.transit import gtfs_realtime_pb2
from google.protobuf.json_format import MessageToDict
from google.protobuf.json_format import MessageToJson
import requests
import json
import pandas as pd
from collections import OrderedDict
import datetime
from protobuf_to_dict import protobuf_to_dict
```

```python
feed = gtfs_realtime_pb2.FeedMessage()
```

```python
feed.ParseFromString(response.content)
for entity in feed.entity:
    if entity.HasField('vehicle'):
        print(entity.vehicle)

dict_obj = protobuf_to_dict(feed)
dict_obj
```

```python
collector = []
counter=0
for block in dict_obj["entity"]:
    counter += 1
    row = OrderedDict()
    row["id"] = block["id"]
    row["trip_id"] = block["vehicle"]["trip"].get("trip_id","")
    row["route_id"] = block["vehicle"]["trip"].get("route_id","")
    row["schedule_relationship"] = block["vehicle"]["trip"].get("schedule_relationship","")
    row["latitude"] = block["vehicle"]["position"].get("latitude","")
    row["longitude"] = block["vehicle"]["position"].get("longitude","")
    row["timestamp"] = block["vehicle"].get("timestamp","")
    row["congestion_level"] = block["vehicle"].get("congestion_level","")
    row["stop_id"] = block["vehicle"].get("stop_id","")
    row["vehicle_id"] = block["vehicle"]["vehicle"].get("id","")
    row["label"] = block["vehicle"]["vehicle"].get("label","")
    collector.append(row)
    df = pd.DataFrame(collector)
```
