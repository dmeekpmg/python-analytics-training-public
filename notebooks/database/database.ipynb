```python
from data.data import Session, path, engine
from pathlib import Path
import zipfile
import pandas as pd

from sqlalchemy import select

from data.model import Shape, Trip
from visualisations.maps import routemap
```

```python
with Session() as session:
    statement = select(Shape).limit(5)
    shapes = session.execute(statement).scalars().all()

shapes
```

```python
with Session() as session:
    stmt = select(Shape).join(Trip).where(Trip.trip_id == 1948209)
    shape = session.execute(stmt).scalars().all()

df = pd.DataFrame.from_records(
    [(s.id, s.lat, s.lon) for s in shape], 
    columns=['shape_id', 'lat', 'lon']
)
```
